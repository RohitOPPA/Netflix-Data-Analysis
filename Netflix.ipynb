Netflix-Data-Analysis

```python
import pandas as pd

df = pd.read_csv('netflix_titles.csv')
df.head()
df['Genre'] =df['listed_in']
```

```python
df.info()
df.describe()
df.columns
```

```python
df.isnull().sum()
df.fillna({'director': 'Unknown'}, inplace=True)
df.fillna({'cast': 'Unknown'}, inplace=True)
df.fillna({'country': 'Unknown'}, inplace=True)
df.dropna(subset=['date_added'], inplace=True)
```

```python
df.drop_duplicates(inplace=True)
```

```python
df['date_added'] = pd.to_datetime(df['date_added'], errors='coerce')
df = df.dropna(subset=['date_added'])
df['year_added'] = df['date_added'].dt.year
df['date_added'].head()
```

```python
df['genre'] = df['listed_in'].apply(lambda x: x.split(',')[0])
df['genre'].head()
```

```python
df['type'].value_counts()
df['country'].value_counts().head(10)
df['cast'].value_counts().head(10)
```

```python
import matplotlib.pyplot as plt

df['type'].value_counts().plot(kind='bar')
plt.title('Distribution of Content Types on Netflix')
plt.xlabel('Content Type')
plt.ylabel('Count')

plt.show()
```

```python
genre_counts = df['genre'].value_counts()
top5 = genre_counts.head(5)

other_count = genre_counts.iloc[5:].sum()

final_data = top5.copy()
final_data['Others'] = other_count

plt.pie(final_data, labels=final_data.index, autopct='%1.1f%%')
plt.title("Genre Distribution (with Others)")
plt.ylabel('')
plt.show()
```

```python
df['year_added'].value_counts().sort_index().plot(kind='line')
plt.title('Number of Titles Added Each Year')
plt.xlabel('Year')
plt.ylabel('Number of Titles')

plt.show()
```

```python
from pandasql import sqldf

# Movies VS TV Shows
sqldf("SELECT type, COUNT(*) FROM df GROUP BY type", globals())
```

```python
# Top 10 Countries
sqldf("SELECT country, COUNT(*) AS Total FROM df GROUP BY country ORDER BY Total DESC LIMIT 10", globals())
```

```python
# Top Genres
sqldf("SELECT Genre, COUNT(*) AS Total FROM df GROUP BY Genre ORDER BY Total DESC", globals())
```

```python
# Content growth over year
sqldf("SELECT year_added, COUNT(*) AS Total FROM df GROUP BY year_added ORDER BY year_added", globals())
```

```python
# Movies VS TV Shows Over Time
sqldf("SELECT year_added, type, COUNT(*) AS Total FROM df GROUP BY year_added, type ORDER BY year_added", globals())
```
